```python
'''
It seems that ChestPainType doesn't positively correlate well with HeartDisease.
So, we might discard that from our features.
Nonetheless, let's address the elephant in the room:
0 min for cholestrol and restingBP.
'''
```

```python
df[df['RestingBP']==0].info()
```

```python
df[df['Cholesterol']==0].info()
```

```python
'''
This is a huge problem, because we don't know yet if cholesterol is a
factor for heart disease or not, and we clearly cannot remove 172 datapoints
from a dataset with 900 datapoints. This is going to require in-depth cleaning.
'''
```

```python
# Let's clean our dataframe.

df_clean = df.copy()
df_clean = df_clean[df_clean['RestingBP'] != 0]
```

```python
'''
As for cholesterol, we're going to replace all the 0 entries with the median,
just so that those values don't positively nor negatively impact the model.
However, in case cholesterol does end up being a factor for HD, we're not going to
use the median for all values; We're going to have two medians, one for the rows with HD
and one for the non-HD rows. If the medians are very different, we won't populate
a bunch of cells with the incorrect median.
So we are going to create a mask and replace all the 0's with the median of the
corresponding values.
'''
hd_mask = df_clean['HeartDisease'] == 0

cholesterol_wo_hd = df_clean.loc[hd_mask, 'Cholesterol'] # Without HD
cholesterol_w_hd = df_clean.loc[~hd_mask, 'Cholesterol'] # With HD

df_clean.loc[hd_mask, 'Cholesterol'] = cholesterol_wo_hd.replace(to_replace = 0, value = cholesterol_wo_hd.median())
df_clean.loc[~hd_mask, 'Cholesterol'] = cholesterol_w_hd.replace(to_replace = 0, value = cholesterol_w_hd.median())
```

```python
df_clean[['Cholesterol', 'RestingBP']].describe()
```

```python
# Our dataset is cleaned!
```

```python
'''
Let's now try to select features for our model. Let's turn all of our features
into numerical values. We will replace the ChestPainType column, which had
values of ATA, NAP, or TA, with three columns for each of those chest pain types,
yielding 1 if the individual has it and 0 otherwise.
'''

df_clean = pd.get_dummies(df_clean, drop_first=True)
df_clean.head()
```

```python
correlations = abs(df_clean.corr()) # Whether we have a positive or negative correlation is not of importance to us, we care about the strength of it
plt.figure(figsize=(12,8))
sns.heatmap(correlations[correlations > 0.30], annot=True, cmap='rocket_r') # 0.30 was chosen to reduce the amount of features to look at on the heat map, but not too many.
plt.show()
```

```python
# We will now finally begin Machine Learning and build our classifier.
```

```python
# splitting train/test sets

X = df_clean.drop(['HeartDisease'], axis=1)
y = df_clean['HeartDisease']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=417) # For reproducibility

features = [
    'MaxHR',
    'Oldpeak',
    'Sex_M',
    'ExerciseAngina_Y',
    'ST_Slope_Flat',
    'ST_Slope_Up'
] # These features survived the 0.30 cutoff from the heat map above

for feature in features:
    knn = KNeighborsClassifier(n_neighbors = 3) # We will play with this hyperparameter later
    knn.fit(X_train[[feature]], y_train)
    accuracy = knn.score(X_val[[feature]], y_val)
    print(f'The k-NN classifier trained on {feature} and with k = 3, has an accuracy of {accuracy*100:.2f}%')
```

```python
'''
The k-NN classifier trained on MaxHR and with k = 3, has an accuracy of 56.52%
The k-NN classifier trained on Oldpeak and with k = 3, has an accuracy of 78.26%
The k-NN classifier trained on Sex_M and with k = 3, has an accuracy of 61.59%
The k-NN classifier trained on ExerciseAngina_Y and with k = 3, has an accuracy of 73.19%
The k-NN classifier trained on ST_Slope_Flat and with k = 3, has an accuracy of 81.88%
The k-NN classifier trained on ST_Slope_Up and with k = 3, has an accuracy of 84.06%

The best accuracy we've got is 84.06%, with the k-NN trained on just one feature,
which is incredible. Let's see if we can beat that. Let's scale the features
and use them all.
'''
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train[features])
X_val_scaled = scaler.transform(X_val[features]) # We don't scale the validation data to a different fit; we transform it on the same list as the training data. 

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train_scaled, y_train)
accuracy = knn.score(X_val_scaled, y_val)
print(f'Accuracy: {accuracy*100:.2f}%')
```

```python
'''
Not as strong as before. That's alright though, we can do hyperparameter optimization.
We will also now start looking at the test sets, since we are concluding this project soon.
'''
```

```python
X = df_clean.drop(['HeartDisease'], axis=1)
y = df_clean['HeartDisease']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 417)

features = [
    'MaxHR',
    'Oldpeak',
    #'Sex_M', # retroactively commented out due to overfitting
    'ExerciseAngina_Y',
    'ST_Slope_Flat',
    'ST_Slope_Up'
]

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train[features])
```

```python
'''
We are now going to use a grid search to test hyperparameters, such as value of k
and distance metric.
'''

grid_params = {'n_neighbors': range(1,20), # 1 to 20 is good for our purpose
               'metric': ['minkowski', 'manhattan']
              }

knn = KNeighborsClassifier()
knn_grid = GridSearchCV(knn, grid_params, scoring='accuracy')
knn_grid.fit(X_train_scaled, y_train)
```

```python
knn_grid.best_score_*100, knn_grid.best_params_ # (82.53846153846153, {'metric': 'manhattan', 'n_neighbors': 19})
```

```python
'''
Now, let's test the model with the test data, with these hyperparamters,
and see what accuracy we get.
'''

X_test_scaled = scaler.transform(X_test[features])
predictions = knn_grid.best_estimator_.predict(X_test_scaled)
accuracy = accuracy_score(y_test, predictions)
print(f'Model Accuracy on test set: {accuracy*100:.2f}%')
```

```python
# Training accuracy: 82.53%, Test accuracy: 82.61%. No under/overfitting!
```

```python
# Let's look at our confusion matrix; Especially important for a healthcare project.

cf = confusion_matrix(y_test, predictions)
ConfusionMatrixDisplay(cf).plot()
plt.show()
```
